```python
import subprocess
import os
from pathlib import Path
```

```python
print(f"当前工作目录: {os.getcwd()}")
notebook_dir = Path.cwd()
print(notebook_dir.parent)
print(os.path.join(notebook_dir.parent,'transfer_learning/transfer_learning.py'))
```

```python
#running transfer learning code for ten adducts 
all_adduct_names = ["2MminH","2MminH2OminH","2MplusH2OplusCl","2MplusKmin2H","2MplusNamin2H","3MplusH2OplusCl","3MplusKmin2H","MminH","MminH2OminH","MplusDANminH"]
notebook_dir = Path.cwd()
for adduct_name in all_adduct_names:
    cmd = [
    'python', os.path.join(notebook_dir.parent,'transfer_learning/transfer_learning.py'),
    '--model_name', 'adduct-trans-0302',
    '--dataset', adduct_name,
    '--best_model_file', os.path.join(notebook_dir.parent,'transfer_learning/model_path/DeepGCN_RT_model.pth'),
    '--type_dataset', 'binary classification',
    '--seed', '1'
        ]   
    process = subprocess.Popen(cmd, stdout=subprocess.PIPE, stderr=subprocess.STDOUT, universal_newlines=True)
    for line in process.stdout:
        print(line, end='')
    process.wait()

print(f"\n✓ finish transfering: {process.returncode}")
```

```python
#get best model from transfer learning result

best_result_getting_code =   os.path.join(notebook_dir.parent,'transfer_learning/get_best_model.py')
script_globals = {}
with open(best_result_getting_code, 'r', encoding='utf-8') as f:
    exec(f.read(), script_globals)
transfer_result = script_globals.get('all_best_final')
```

```python
#get ECFP,GCN,mix_fp
for adduct_name in all_adduct_names:
    model_name_temp =  os.listdir(os.path.join(notebook_dir.parent,'transfer_learning/model_path/best',adduct_name))[0]
    cmd = [
    'python', os.path.join(notebook_dir.parent,'predictors/get_mix_fp.py'),
    '--dataset', adduct_name,
    '--best_model_file', os.path.join(notebook_dir.parent,'transfer_learning/model_path/best',adduct_name,model_name_temp),
    '--all_fp_path',  os.path.join(notebook_dir.parent,'predictors/FPs',adduct_name)
        ]   
    process = subprocess.Popen(cmd, stdout=subprocess.PIPE, stderr=subprocess.STDOUT, universal_newlines=True)
    for line in process.stdout:
        print(line, end='')
    process.wait()
    
```

```python
#predictor part
for adduct_name in all_adduct_names:
    cmd = [
    'python', os.path.join(notebook_dir.parent,'predictors/baseline_ECFP.py'),
    '--name', adduct_name
        ] 
    process = subprocess.Popen(cmd, stdout=subprocess.PIPE, stderr=subprocess.STDOUT, universal_newlines=True)
    for line in process.stdout:
        print(line, end='')
    process.wait()

    cmd = [
    'python', os.path.join(notebook_dir.parent,'predictors/baseline_GCN_FP.py'),
    '--name', adduct_name
        ] 
    process = subprocess.Popen(cmd, stdout=subprocess.PIPE, stderr=subprocess.STDOUT, universal_newlines=True)
    for line in process.stdout:
        print(line, end='')
    process.wait()

    cmd = [
    'python', os.path.join(notebook_dir.parent,'predictors/baseline_mix.py'),
    '--name', adduct_name
        ] 
    process = subprocess.Popen(cmd, stdout=subprocess.PIPE, stderr=subprocess.STDOUT, universal_newlines=True)
    for line in process.stdout:
        print(line, end='')
    process.wait()
```
